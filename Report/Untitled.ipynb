```python
import numpy as np

ak= lambda k: 1/4*(2*((1j*np.pi*k+1)*(-1)**k-1)/(np.pi**2*k**2)+2*(1j-1j*(-1)**k)/(np.pi*k))
```

```python
ak=lambda k: -1/(2*np.pi**2*k**2)+(-1)**k/(2*np.pi**2*k**2)+(-1)**k/(2*np.pi*k*1j)-1/(2*np.pi*k*1j)
```

```python
t=np.arange(-8,8,0.001)
y=np.zeros(t.shape)
```

```python
for k in range(-50,50):
    if (k!=0):
        c=ak(k)
    else:
        c=0.75
        
    y+=np.real(c*np.exp(1j*k*np.pi/2*t))
        
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t,y)
```
